```python
import numpy as np
import pandas as pd
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import LabelEncoder , OneHotEncoder
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
```

```python
df = pd.DataFrame({
    'B': [1,3.5,np.nan,8,np.nan,np.nan,9.8],
    'C':['yellow',np.nan,'red','black','red',np.nan,'white'],
    'D':[np.nan,700,100,222,550,np.nan,200],
    'target':['dog','cat','cat','dog','cat','dog','dog'] 
})
df
```

```python
df.isna().any().any()
```

```python
num_imputer = SimpleImputer(strategy='mean')
cat_imputer = SimpleImputer(strategy='most_frequent')

preprocessor = ColumnTransformer(transformers=[
    ('num' , num_imputer , ["B","D"]),
    ('cat', cat_imputer , ["C"])
])

df[["B","D","C"]] = preprocessor.fit_transform(df[["B","D","C"]])
df
```

```python
encoder = OneHotEncoder(sparse=False)
encoder_features = encoder.fit_transform(df[["C","target"]])
features_names = encoder.get_feature_names_out(["C","target"])
df_encoder = pd.DataFrame(encoder_features, columns=features_names)
df_final = pd.concat([df.drop(['C','target'], axis=1),df_encoder],axis=1)

x = df_final.drop(['target_dog','target_cat'], axis=1)
y = df_final[['target_dog','target_cat']]
df_final
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(x,y,test_size=0.3, random_state=42)
X_train
```

```python
X_test
```

```python
sc = StandardScaler()
X_train_sc = sc.fit_transform(X_train)
X_test_sc = sc.transform(X_test)
```
